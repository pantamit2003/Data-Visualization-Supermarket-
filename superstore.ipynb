# Exploratory Data Analysis

```python
df.columns
```

### Profit based on product category

```python
sales_by_region = df.groupby('Product Category')['Profit'].sum().reset_index()

x = sales_by_region['Product Category']
y = sales_by_region['Profit'] 

plt.bar(x,y)
plt.xlabel("Product category") 
plt.ylabel("profit") 
plt.title("Profit Based category")
plt.show()
```

```python
df.columns
```

### Profit based on Unit Price

```python
x=df['Unit Price']
y=df['Profit']

plt.scatter(x,y)
plt.xlabel('Unit Price')
plt.ylabel('Profit') 
plt.title("SCATTER PLOT",fontsize=12)

plt.show()
```

### Region wise Profit

```python


sales_by_region = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

x = sales_by_region['Region']
y_sales = sales_by_region['Sales']
y_profit = sales_by_region['Profit']

plt.figure(figsize=(10, 6))
plt.stackplot(x, y_sales, y_profit, labels=['Sales', 'Profit'])
plt.legend(loc='upper left')
plt.xlabel("Region")
plt.ylabel("Amount")
plt.title("Sales and Profit by Region")
plt.show()
```

## Sales Distribution by Product Container

```python

sales_by_container = df.groupby('Product Container')['Sales'].sum().reset_index()

x = sales_by_container['Sales']
y = sales_by_container['Product Container']

plt.pie(x, labels=y, autopct='%1.1f%%')
plt.title('Sales Distribution by Product Container')
plt.show()
```

## Profit Distribution based on Product Container

```python
sales_by_container = df.groupby('Product Container')['Profit'].sum().reset_index()

x = sales_by_container['Product Container']
y = sales_by_container['Profit']

plt.figure(figsize=(12, 6))
plt.bar(x, y, width=0.5) 
plt.xlabel("Product Container")
plt.ylabel("Profit")
plt.title("Profit Based Product Container")
plt.xticks(rotation=45)  
plt.show()
```

## Total sales by product Category

```python


sales_by_region = df.groupby('Product Category')['Sales'].sum().reset_index()

x = sales_by_region['Product Category']
y = sales_by_region['Sales']

plt.plot(x, y)
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Total Sales by Product Category")
 
plt.show()


```
